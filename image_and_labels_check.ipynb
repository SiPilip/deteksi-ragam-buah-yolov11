```python
import os

# Folder label dataset
source_label_folders = [
    'dataset/apel/train/labels/',
    'dataset/belimbing/train/labels/',
    'dataset/belimbing2/train/labels/',
    'dataset/ceri/train/labels/',
    'dataset/kiwi/train/labels/',
    'dataset/kiwi2/train/labels/',
    'dataset/kurma/train/labels/',
    'dataset/lemon/train/labels/',
    'dataset/nanas/train/labels/',
    'dataset/pisang/train/labels/',
    'dataset/semangka/train/labels/',
    'dataset/tomat/train/labels/'
]

# Folder dataset
source_image_folders = [
    'dataset/apel/train/images/',
    'dataset/belimbing/train/images/',
    'dataset/belimbing2/train/images/',
    'dataset/ceri/train/images/',
    'dataset/kiwi/train/images/',
    'dataset/kiwi2/train/images/',
    'dataset/kurma/train/images/',
    'dataset/lemon/train/images/',
    'dataset/nanas/train/images/',
    'dataset/pisang/train/images/',
    'dataset/semangka/train/images/',
    'dataset/tomat/train/images/'
]


labels_count = []
images_count = []

count = 0

for folder in source_label_folders:
  for path in os.listdir(folder):
     count += 1
  labels_count.append(count)
  count = 0

for folder in source_image_folders:
  for path in os.listdir(folder):
      count += 1
  images_count.append(count)
  count = 0

print(f"Array jumlah label: {labels_count}")
print(f"Array jumlah gambar: {images_count}")
```
