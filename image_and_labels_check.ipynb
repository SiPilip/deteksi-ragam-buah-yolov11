## Pengecekan jumlah data gambar dan labels pada dataset


```python
import os

# Folder label dataset
source_label_folders = [
    'data_reduction/apel/train/labels/',
    'data_reduction/belimbing/train/labels/',
    'data_reduction/ceri/train/labels/',
    'data_reduction/kiwi/train/labels/',
    'data_reduction/kurma/train/labels/',
    'data_reduction/lemon/train/labels/',
    'data_reduction/nanas/train/labels/',
    'data_reduction/pisang/train/labels/',
    'data_reduction/semangka/train/labels/',
    'data_reduction/tomat/train/labels/'
]

# Folder dataset
source_image_folders = [
    'data_reduction/apel/train/images/',
    'data_reduction/belimbing/train/images/',
    'data_reduction/ceri/train/images/',
    'data_reduction/kiwi/train/images/',
    'data_reduction/kurma/train/images/',
    'data_reduction/lemon/train/images/',
    'data_reduction/nanas/train/images/',
    'data_reduction/pisang/train/images/',
    'data_reduction/semangka/train/images/',
    'data_reduction/tomat/train/images/'
]


labels_count = []
images_count = []

count = 0

for folder in source_label_folders:
  for path in os.listdir(folder):
     count += 1
  labels_count.append(count)
  count = 0

for folder in source_image_folders:
  for path in os.listdir(folder):
      count += 1
  images_count.append(count)
  count = 0

print(f"Array jumlah label: {labels_count}")
print(f"Array jumlah gambar: {images_count}")
```
