## Pengggabungan Data

Dikarenakan kita memiliki data gambar dari 10 kelas yang berbeda beda.

Dibutuhkan fungsi untuk menggabungkan tiap dataset gambar dari 10 kelas, menjadi satu dataset utuh yang akan digunakan oleh model.


```python
import os
import shutil

# Folder dataset
source_folders = [
    'data_reduction/apel/train/images',
    'data_reduction/belimbing/train/images/',
    'data_reduction/ceri/train/images/',
    'data_reduction/kiwi/train/images/',
    'data_reduction/kurma/train/images/',
    'data_reduction/lemon/train/images/',
    'data_reduction/nanas/train/images/',
    'data_reduction/pisang/train/images/',
    'data_reduction/semangka/train/images/',
    'data_reduction/tomat/train/images/'
]

# Folder tujuan
output_folder = './data_images_for_train_2/images'
os.makedirs(output_folder, exist_ok=True)

totalDataTrain = 0

for folder in source_folders: # tiap folder dalam folder yang kita definisike
  for filename in os.listdir(folder): # tiap file dalam folderny
    source_path = os.path.join(folder, filename) # ex. dataset/apel/train/images
    destination_path = os.path.join(output_folder, filename) # ex. images/kiwi10330.jpg
    shutil.copy2(source_path, destination_path)
    totalDataTrain += 1

print(f'{totalDataTrain} gambar udah berhasil digabung! ✅')
```
