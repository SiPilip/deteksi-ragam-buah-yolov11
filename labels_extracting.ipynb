## Ekstraksi Nama Kelas

Dikarenakan kita memiliki data 10 kelas yang berbeda beda dengan format YOLOv11.

Dibutuhkan fungsi untuk mengekstraksi tiap nama kelas dalam folder labels lalu kita sesuaikan menjadi nama kelas yang sudah kita tetapkan untuk tiap kelas dalam model kita


```python
import os

# Map kelas ke indeks
class_mapping = {
    'apel': 0,
    'belimbing': 1,
    'belimbing2': 1,
    'ceri': 2,
    'kiwi': 3,
    'kiwi2': 3,
    'kurma': 4,
    'lemon': 5,
    'nanas': 6,
    'pisang': 7,
    'semangka': 8,
    'tomat': 9
}

# Folder dataset
source_folders = [
    'dataset/apel/train/labels/',
    'dataset/belimbing/train/labels/',
    'dataset/belimbing2/train/labels/',
    'dataset/ceri/train/labels/',
    'dataset/kiwi/train/labels/',
    'dataset/kiwi2/train/labels/',
    'dataset/kurma/train/labels/',
    'dataset/lemon/train/labels/',
    'dataset/nanas/train/labels/',
    'dataset/pisang/train/labels/',
    'dataset/semangka/train/labels/',
    'dataset/tomat/train/labels/'
]

# Folder tujuan
output_folder = 'labels/'
os.makedirs(output_folder, exist_ok=True)

totalLabelDataTrain = 0

for folder in source_folders:
    fruit = folder.split('/')[-4]  # Ambil nama buah dari path
    class_id = class_mapping[fruit]  # Dapatkan ID kelas yang sesuai

    for filename in os.listdir(folder):
        if filename.endswith('.txt'):
            input_path = os.path.join(folder, filename)
            output_path = os.path.join(output_folder, filename)

            with open(input_path, 'r') as f:
                lines = f.readlines()

            new_lines = []
            for line in lines:
                parts = line.split()
                parts[0] = str(class_id)  # Ganti ID kelas dengan yang benar
                new_lines.append(' '.join(parts))

            with open(output_path, 'w') as f:
                f.write('\n'.join(new_lines))

            totalLabelDataTrain += 1
            
print(f'{totalLabelDataTrain} label berhasil digabung dan diperbarui! ✅')
```
