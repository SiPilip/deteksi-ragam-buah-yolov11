```python
import os
import shutil

nama_kelas= "kiwi"

output_images_folder = f"data_reduction/{nama_kelas}/train/images"
os.makedirs(output_images_folder, exist_ok=True)

output_labels_folder = f"data_reduction/{nama_kelas}/train/labels"
os.makedirs(output_labels_folder, exist_ok=True)

source_images_folder = f"./__DATASET - Copy/{nama_kelas}/train/images"
source_labels_folder = f"./__DATASET - Copy/{nama_kelas}/train/labels"

reduction_limit = 500 # tentukan berapa jumlah data gambar yang akan kita reduksi dari dataset awal

count = 0

# Reduksi gambar
for file in os.listdir(source_images_folder):

  if count == len(os.listdir(source_images_folder))-1:
    print(f"Tidak dapat mereduksi karena data gambar kurang dari {reduction_limit}, namun operasi tetap berhasil dijalankan")
    break

  if count == reduction_limit:
    print(f"Berhasil mereduksi sebanyak {len(os.listdir(source_images_folder))} data gambar menjadi {reduction_limit} data gambar")
    break
  
  source_path = os.path.join(source_images_folder, file) # ex. dataset/apel/train/images
  destination_path = os.path.join(output_images_folder, file) # ex. images/kiwi10330.jpg
  shutil.copy2(source_path, destination_path)

  count += 1

count = 0

# Reduksi gambar
for file in os.listdir(source_labels_folder):

  if count == len(os.listdir(source_labels_folder))-1:
    print(f"Tidak dapat mereduksi karena data label kurang dari {reduction_limit}, namun operasi tetap berhasil dijalankan")
    break

  if count == reduction_limit:
    print(f"Berhasil mereduksi sebanyak {len(os.listdir(source_labels_folder))} data label menjadi {reduction_limit} data label")
    break
  
  source_path = os.path.join(source_labels_folder, file) # ex. dataset/apel/train/images
  destination_path = os.path.join(output_labels_folder, file) # ex. images/kiwi10330.jpg
  shutil.copy2(source_path, destination_path)

  count += 1
```
